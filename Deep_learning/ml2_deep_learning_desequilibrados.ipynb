En primer lugar se cargan las librerías que se van a necesitar:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
from skimage.transform import resize
from random import sample, seed
import tensorflow as tf
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

A continuación, se cargan los datos. Se disponen de dos carpetas, una de ellas para cada tipo de célula (infectadas y no infectadas). 

```python
ruta_parasited = "cell_images/Parasitized/"
ruta_uninfected = "cell_images/Uninfected/"

parasitized = os.listdir(ruta_parasited) 
uninfected = os.listdir(ruta_uninfected)
```

Se realiza una visualización de 4 células con parásito y 4 sin parásito. Se aprecia claramente, que las células con parásito muestran una zona más oscura, identificando así al parásito.

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(10,10))
for i in range(4):
    image = parasitized[i]
    imagen1 = plt.imread(ruta_parasited + image)
    ax = axes[0,i]
    ax.imshow(imagen1)

for i in range(4):
    image = uninfected[i]
    imagen1 = plt.imread(ruta_uninfected + image)
    ax = axes[1,i]
    ax.imshow(imagen1)
```

En segundo lugar, se leen las imágenes mediante la función `imread` de `matplotlib` de tal forma que se obtendrán dos listas (células infectadas y no infectadas) con todas las imágenes de cada tipo con los píxeles de ancho x alto x 3 capas (imagen a color creada mediante `RGB`). Además, se seleccionan 2500 imágenes de cada tipo por problemas de memoria.

```python
lista_parasited = []
lista_uninfected = []
tamaño_fila_u = []
tamaño_colum_u = []
tamaño_fila_p = []
tamaño_colum_p = []

parasitized = parasitized[0:2500]
uninfected = uninfected[0:2500]

for image_p in parasitized:
    imagen1_p = plt.imread(ruta_parasited + image_p)
    lista_parasited.append(imagen1_p)
    tamaño_fila_p.append(imagen1_p.shape[0])
    tamaño_colum_p.append(imagen1_p.shape[1])

for image_u in uninfected:
    imagen1_u = plt.imread(ruta_uninfected + image_u)
    lista_uninfected.append(imagen1_u)
    tamaño_fila_u.append(imagen1_u.shape[0])
    tamaño_colum_u.append(imagen1_u.shape[1])
    
```

Por ejemplo, la primera imagen en la lista de células infectadas, tendría 148 píxeles de alto (número de filas), y 142 de ancho (número de columnas) y 3 capas (RGB, para obtener la imagen a color):

```python
lista_parasited[0].shape
```

Se ha observado que no todas las imágenes tienen el mismo número de filas y columnas, por lo que se hace preciso realizar una redimensión para que todas las células tengan el mismo tamaño. Para realizar este redimensionamiento, primero se busca la célula con un mayor ancho o alto (`tamaño_max_total`) y se adaptan todas las imágenes a éste: `tamaño_max_total*tamaño_max_total`.

```python
tamaño_max_filas = max(max(tamaño_fila_u),max(tamaño_fila_p))
tamaño_max_colum = max(max(tamaño_colum_u),max(tamaño_colum_p))
tamaño_max_total = max(tamaño_max_filas,tamaño_max_colum)

print("Tamaño max filas: ", tamaño_max_filas)
print("Tamaño max columnas: ", tamaño_max_colum)
print("Tamaño max total: ", tamaño_max_total)

list_resize_p = []
list_resize_u = []

for image_p in lista_parasited:
    list_resize_p.append(resize(image_p,(tamaño_max_total,tamaño_max_total)))

for image_u in lista_uninfected:
    list_resize_u.append(resize(image_u,(tamaño_max_total,tamaño_max_total)))
    
```

Se vuelven a mostrar 4 ejemplos de imágenes con y sin parásito para comprobar el redimensionado:

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(10,10))
for i in range(4):
    image = list_resize_p[i]
    ax = axes[0,i]
    ax.imshow(image)

for i in range(4):
    image = list_resize_u[i]
    ax = axes[1,i]
    ax.imshow(image)
```

Como el tamaño actual por imagen es de 238x238, se decide bajar la resolución para optimizar el rendimiento de la red, para ello se realiza un `diezmado`.

```python
resPixel=50
pixeles = np.linspace(0, tamaño_max_total -1, num=resPixel)

indDown=[]
for k_fila in pixeles:
    for k_colum in pixeles:
        indDown.append(int(k_fila)*tamaño_max_total+int(k_colum))
        
list_red_p = []
list_red_u = []

for image in list_resize_p:
    imagen = np.concatenate(image)
    imagen_1 = imagen[indDown,:]
    list_red_p.append(imagen_1.reshape(resPixel,resPixel,3))

for image in list_resize_u:
    imagen = np.concatenate(image)
    imagen_1 = imagen[indDown,:]
    list_red_u.append(imagen_1.reshape(resPixel,resPixel,3))
```

A continuación se vuelven a mostrar las imágenes diezmadas, y se comprueba que a pesar del diezmado, se sigue apreciando la mancha que identifica a las células infectadas.

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(10,10))
for i in range(4):
    image = list_red_p[i]
    ax = axes[0,i]
    ax.imshow(image)

for i in range(4):
    image = list_red_u[i]
    ax = axes[1,i]
    ax.imshow(image)
```

# 1.4. Train, test y validación

Se dividen las imágenes en tres conjuntos: `train` (70%), `test` (15%) y `validación` (15%) para cada tipo de célula.

```python
# parasited
num_total_p=np.arange(len(list_red_p))
seed(122556)

# 70% para train
indices_train_p = np.array(sample(list(num_total_p), int(0.7*len(num_total_p))))
list_red_p_array = np.array(list_red_p)
XTrain_p = list_red_p_array[indices_train_p]

# 15% para test
indices_p=np.arange(len(num_total_p))
indices_test_p=np.setdiff1d(indices_p, indices_train_p)

indices_test1_p = np.array(sample(list(indices_test_p), int(0.15*len(num_total_p))))
XTest_p = list_red_p_array[indices_test1_p]

# 15% para validacion
indices_validacion_p=np.setdiff1d(indices_p, indices_train_p) 
indices_validacion_p=np.setdiff1d(indices_validacion_p,indices_test1_p)
XVal_p=list_red_p_array[indices_validacion_p]

print("Número de muestras infectadas en train: ", len(XTrain_p))
print("Número de muestras infectadas en test: ", len(XTest_p))
print("Número de muestras infectadas en validación: ", len(XVal_p))
```

```python
# uninfected
num_total_u=np.arange(len(list_red_u))
seed(122556)

# 70% para train
indices_train_u = np.array(sample(list(num_total_u), int(0.7*len(num_total_u))))
list_red_u_array = np.array(list_red_u)
XTrain_u = list_red_u_array[indices_train_u]

# 15% para test
indices_u=np.arange(len(num_total_u))
indices_test_u=np.setdiff1d(indices_u, indices_train_u)

indices_test1_u = np.array(sample(list(indices_test_u), int(0.15*len(num_total_u))))
XTest_u = list_red_u_array[indices_test1_u]

# 15% para validacion
indices_validacion_u=np.setdiff1d(indices_u, indices_train_u) 
indices_validacion_u=np.setdiff1d(indices_validacion_u,indices_test1_u)
XVal_u=list_red_u_array[indices_validacion_u]

print("Número de muestras no infectadas en train: ", len(XTrain_u))
print("Número de muestras no infectadas en test: ", len(XTest_u))
print("Número de muestras no infectadas en validación: ", len(XVal_u))
```

Se junta en un único conjunto de train, test y validación las células infectadas y no infectadas, y se crean las etiquetas para cada célula que serán los `outputs` de la red neuronal. Se identificará con `etiqueta 0` a las células infectadas y con `etiqueta 1` a las no infectadas.

```python
XTrain_previo = np.concatenate((XTrain_p, XTrain_u))
XTrain = []

for i in XTrain_previo:
    XTrain.append(np.concatenate(i))

XTrain = np.array(XTrain)

YTrain_p = np.zeros(len(XTrain_p),dtype=int)
YTrain_u = np.ones(len(XTrain_u),dtype=int)
YTrain = np.concatenate((YTrain_p, YTrain_u))

print("Número de muestras totales en train: ", len(XTrain))
print("Número de etiquetas en train: ", len(YTrain))
```

```python
XTest_previo = np.concatenate((XTest_p, XTest_u))
XTest = []

for i in XTest_previo:
    XTest.append(np.concatenate(i))

XTest = np.array(XTest)

YTest_p = np.zeros(len(XTest_p),dtype=int)
YTest_u = np.ones(len(XTest_u),dtype=int)
YTest = np.concatenate((YTest_p, YTest_u))

print("Número de muestras totales en test: ", len(XTest))
print("Número de etiquetas en test: ", len(YTest))
```

```python
XVal_previo = np.concatenate((XVal_p, XVal_u))
XVal = []

for i in XVal_previo:
    XVal.append(np.concatenate(i))

XVal = np.array(XVal)

YVal_p = np.zeros(len(XVal_p),dtype=int)
YVal_u = np.ones(len(XVal_u),dtype=int)
YVal = np.concatenate((YVal_p, YVal_u))

print("Número de muestras totales en validación: ", len(XVal))
print("Número de etiquetas en validación: ", len(YVal))
```

Se reestructuran los outputs, para que en lugar de ser una lista de ceros y unos, sean una tupla, donde se active la posición correspondiente al tipo de célula. La `tupla` tendrá la primera posición activada si la `etiqueta` es `cero` [1,0], y tendrá la segunda posición activada si tiene `etiqueta uno` [0,1].

```python
# Para train

y_new_train=[]

for i in range(len(YTrain)):

    if YTrain[i] == 0:
        y_new_train.append([1,0])
    else:
        y_new_train.append([0,1])

        
y_new_train = np.array(y_new_train) 
print('Estructura de las etiquetas de train: ', y_new_train.shape)

# Para Test

y_new_test=[]

for i in range(len(YTest)):

    if YTest[i] == 0:
        y_new_test.append([1,0])
    else:
        y_new_test.append([0,1])

        
y_new_test = np.array(y_new_test) 
print('Estructura de las etiquetas de test: ', y_new_test.shape)

# Para validacion

y_new_val=[]

for i in range(len(YVal)):

    if YVal[i] == 0:
        y_new_val.append([1,0])
    else:
        y_new_val.append([0,1])

        
y_new_val = np.array(y_new_val) 
print('Estructura de las etiquetas de validación: ', y_new_val.shape)
```

# 1.5. Implementación del modelo: red neuronal

En primer lugar, para la implementación de la red neuronal, se configuran algunos parámetros.

```python
tf.keras.backend.clear_session()

# Parámetros generales de la red en función de la resolución
N_capas = 3 # Número de etapas para la red neuronal
N_tags = 2  # Número de clases (infected, uninfected)

Nepoch=5 # Número de épocas
Nbatch=3 # Número de muestras de cada mini-batch

pDO=0.3 # Probabilidad de dropout

Npix=50    # Número de píxeles por dirección        
Nconv1=3   # Tamaño del filtro convolucional 1 (píxeles)
Nconv2=3   # Tamaño del filtro convolucional 2 (píxeles) 
Nfil1=16   # Número de filtros de la primera etapa
Nfil2=32   # Número de filtros de la segunda etapa
Nneu=512   # Número de neuronas del MLP final
```

Se crea la red neuronal con las siguientes capas: `Reshape`, `Conv2D`, `MaxPooling2D`, `Conv2D`, `Flatten`, `Dense`, `Dropout` y `Dense`.

```python
# Se define la estructura básica de la red CNN usando TensorFlow y Keras

tf.keras.backend.clear_session()

model = tf.keras.Sequential(
  [
      tf.keras.layers.Reshape(input_shape=(Npix*Npix,N_capas), target_shape=(Npix, Npix,N_capas)),
      tf.keras.layers.Conv2D(kernel_size=Nconv1, filters=Nfil1, activation='relu', padding='same'),
      tf.keras.layers.MaxPooling2D((2, 2)),
      tf.keras.layers.Conv2D(kernel_size=Nconv2, filters=Nfil2, activation='relu', padding='same'),
      tf.keras.layers.Flatten(), #capa para ajustar el numero de elementos
      tf.keras.layers.Dense(Nneu, activation='relu'), #capa de activación
      tf.keras.layers.Dropout(pDO),
      tf.keras.layers.Dense(N_tags, activation='softmax')
  ])
    
model.compile(optimizer="rmsprop",
              loss='categorical_crossentropy',
              metrics=['accuracy'])
    
model.summary()
```

Se usa el modelo implementado con los datos de nuestro experimento, y se comparan las etiquetas predichas con las reales. Se obtiene un `accuracy` de 80.6% para el conjunto de train y 76.67% para el conjunto de test.

```python
model.fit(XTrain,y_new_train,
          batch_size=Nbatch,
          epochs=Nepoch,
          initial_epoch=0
         )  

oTrain=model.predict(XTrain)
oTest=model.predict(XTest)
        
YTrainE=oTrain.argmax(axis=1)
YTestE=oTest.argmax(axis=1)

PaTrain=100*accuracy_score(YTrain,YTrainE)
PaTest=100*accuracy_score(YTest,YTestE)
        
print('Probabilidad de acierto (Train): %2.2f %%'%(PaTrain))
print('Probabilidad de acierto (Test): %2.2f %%'%(PaTest))
```

# 1.6. Evaluación del modelo:

Por último, para evaluar el modelo, se calcula la `matriz de confusión`, se muestra el  `informe` con el resultado de la `clasificación` y el `accuracy` para el conjunto de test.

```python
print('Matriz de confusión: \n', confusion_matrix(YTest , YTestE))
```

De la matriz de confusión, se concluye que el modelo acierta un total de 684 muestras (entre las infectadas y no infectadas) de las 750 totales y falla en 66 muestras. 

```python
print('Informe de clasificación: \n', classification_report(YTest , YTestE))
```

Con el informe de clasificación se aprecia que la **precisión** de la etiqueta 0 (infectadas) es de 0.82 y para la 1 (no infectadas) es de 0.73. Si se analiza la **recall**, en este caso es superior para la etiqueta 1 (0.86) que para la etiqueta 0 (0.68). La **f1-medida** también tiene buenos resultados para ambas etiquetas (0.765). Por último, como ya se había calculado, el **accuracy** es de 0.77 para el conjunto de test.

Una vez analizadas estas métricas, se concluye que el modelo funciona bastante bien con las 2.500 muestras consideradas (del total de 13.779 con las que se contaba de cada tipo) y con el modelo implementado.

# 2.5. Implementación de modelos

## 2.5.1. Implementación de red neuronal con coste de Bregman

Se define la red neuronal MLP que únicamente tendrá una capa oculta, cuyos parámetros serán el número de neuronas de la capa oculta `M`, la tasa de aprendizaje `lr` y la función de activación `activation`. El coste subrogado será el MSE (coste de Bregman).

```python
def make_model(M, lr, activation):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(M, activation=activation, input_dim=X_train.shape[1]))
    model.add(tf.keras.layers.Dense(1, activation=activation))
    
    model.compile(
        optimizer=keras.optimizers.RMSprop(lr=lr),
        loss=keras.losses.MeanSquaredError() # coste bregman
    )
    
    return model

M = 10
lr = 0.001

activation = 'tanh'
epochs=300
batch_size=256

model_bregman = make_model(M, lr, activation)
model_bregman.summary()
```

Se aplica la red neuronal implementada sobre los datos y se realiza la predicción para el conjunto de test.

```python
history = model_bregman.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)
y_pred_test = model_bregman.predict(X_test).ravel()
```

## 2.5.2. Implementación con sobre-muestreo "neutral": SMOTE

A continuación, se utiliza una técnica de sobre-muestreo `SMOTE` para generar muestras artificiales de la clase minoritaria.

```python
sm = SMOTE(random_state=50)
X_smo, y_smo = sm.fit_resample(X_train, y_train)
```

Como se comprueba a continuación, después de aplicar el sobre-muestreo SMOTE se tienen muestras balanceadas con 8013 observaciones de la clase mayoritaria (casas baratas) y 8013 de la minoritaria (casas caras).

```python
n_minority_smo = np.sum(y_smo == 1)
n_majority_smo = np.sum(y_smo == -1)

IR_smo = n_majority_smo/n_minority_smo

print('Número de casas caras: '+ str(n_minority_smo))
print('Número de casas baratas: '+ str(n_majority_smo))
print('Ratio de desequilibrio: '+ str(IR_smo))
```

Se entrena el mismo modelo que en el caso anterior, pero ahora usando los datos balanceados creados en esta sección: `X_smo` e `y_smo`. 

```python
model_smo = make_model(M, lr, activation)
model_smo.summary()
```

```python
history_smo = model_smo.fit(X_smo, y_smo, epochs=epochs, batch_size=batch_size, verbose=0)
y_pred_test_smo = model_smo.predict(X_test).ravel()
```

## 2.5.3. Implementación con sobre-muestreo "informado": Borderline-SMOTE

A continuación, se usa un tipo diferente de SMOTE, SMOTE-Borderline. Este tipo de SMOTE, genera muestras sintéticas a partir de semillas de la clase minoritaria cercanas al borde.

```python
sm_borderline = BorderlineSMOTE(random_state=42)
X_borderline_smo, y_borderline_smo = sm_borderline.fit_resample(X_train, y_train)
```

```python
n_minority_smo = np.sum(y_borderline_smo == 1)
n_majority_smo = np.sum(y_borderline_smo == -1)

IR_smo = n_majority_smo/n_minority_smo

print('Número de casas caras: '+ str(n_minority_smo))
print('Número de casas baratas: '+ str(n_majority_smo))
print('Ratio de desequilibrio: '+ str(IR_smo))
```

```python
model_borderline_smo = make_model(M, lr, activation)
model_borderline_smo.summary()
```

Se entrena el modelo con las nuevas muestras creadas con SMOTE-Borderline:

```python
history_borderline_smo = model_borderline_smo.fit(X_borderline_smo, 
                                                  y_borderline_smo, 
                                                  epochs=epochs, 
                                                  batch_size=batch_size, 
                                                  verbose=0)
y_pred_test_borderline_smo = model_borderline_smo.predict(X_test).ravel()
```

# 2.6. Evaluación de los modelos:

Por último, se evalúan y comparan los tres modelos creados para datos desequilibrados. En primer lugar, se muestran las `curvas ROC` de cada modelo y después, la `matriz de confusión`, el `informe de clasificación` y el `accuracy`.

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_test, pos_label=1)
fpr_smo, tpr_smo, thresholds_smo = metrics.roc_curve(y_test, y_pred_test_smo, pos_label=1)
fpr_bsmo, tpr_bsmo, thresholds_bsmo = metrics.roc_curve(y_test, y_pred_test_borderline_smo, pos_label=1)
```

```python
plt.figure(figsize = (7, 6))
plt.axis([0, 1, 0, 1])
plt.plot(fpr, tpr, fpr_smo, tpr_smo, fpr_bsmo, tpr_bsmo)
plt.xlabel('$P_{FA}$', fontsize = 15)
plt.ylabel('$P_D$', fontsize = 15)
plt.legend(('MLP-coste Bregman','SMOTE','B-SMOTE'),loc='lower right')
plt.show()
```

Con un muestreo SMOTE, se obtiene un **accuracy** de 0.87. La **precisión** para cada clase es de 0.95 y 0.68, respectivamente y la recall de 0.87 para la clase -1 y 0.86 para la clase 1. En este modelo empeora la clasificación de la clase 1 (clase minoritaria) respecto al modelo anterior.
